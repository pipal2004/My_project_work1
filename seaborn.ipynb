```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30, palette='Set1')
plt.title("Age Distribution by Survival", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=titanic, x='pclass', hue='survived', palette='coolwarm')
plt.title("Survival Count by Class", fontsize=14, fontweight='bold')
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x='sex', hue='survived', palette='pastel')
plt.title("Survival Count by Gender", fontsize=14, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='survived', y='age')
plt.title("Age Distribution by Survival (Boxplot)", fontsize=14, fontweight='bold')
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_cols = titanic.select_dtypes(include='number')
corr = numeric_cols.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Columns)", fontsize=14, fontweight='bold')
plt.show()
```

```python
sns.pairplot(titanic, hue='survived', vars=['age','fare'], palette='Set1')
plt.show()
```
