```python
#Create a sample dataset for Line plot
import matplotlib.pyplot as plt
students = ["A", "B", "C", "D", "E", "F", "G"]
marks = [78, 85, 69, 92, 74, 88, 95]
plt.plot(students,marks,marker='o',color='b',label="Marks Trend")
plt.title("Line Plot: Student Marks", fontsize=14, fontweight='bold')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
#Create a sample dataset for Bar plot
plt.bar(students, marks, color='green', label="Marks")
plt.title("Bar Plot: Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
#Create a sample dataset for Scatter  plot
plt.scatter(students, marks, color='red', marker='*', s=100, label="Marks")
plt.title("Scatter Plot: Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
